## LOC

```python
# Pegando dados da primeira linha
dados.head(1)
```

- Esse m√©todo possui dois par√¢metros:
  - **N√∫mero de linhas:** passado da seguinte forma ``[ValorInicial:ValorFinal]``;
  - **Nome das colunas:** passado via lista

```python
# loc[Linhas, ['NomeColuna']]
dados.loc[20:40, ['Survived', 'Name', 'Sex', 'Age']]
```

```python
# Pode tamb√©m s√≥ informar o valor inicial e n√£o informar o valor final no primeiro par√¢metro
# A biblioteca Pandas entender√° que voc√™ quer do valor informado at√© o valor final
dados.loc[800: , ['Survived', 'Name', 'Sex', 'Age']]
```

```python
# Tamb√©m posso n√£o informar o valor inicial e informar somente o valor final
# A biblioteca Pandas entender√° que voc√™ quer do valor inicial at√© o valor informado de linhas
dados.loc[ :500, ['Survived', 'Name', 'Sex', 'Age']]
```

```python
# Uma boa pr√°tica: criar uma vari√°vel com a nossa consulta
dados2 = dados.loc[ :500, ['Survived', 'Name', 'Sex', 'Age']]
```

## ILOC

- Esse m√©todo possui dois par√¢metros:
  - **N√∫mero de linhas:** passado da seguinte forma ``[ValorInicial:ValorFinal]``;
  - **√çndice de colunas:** passado da seguinte forma ``[√çndiceColunaInformada:√çndiceColunaFinal+1]``:
    - O √çndice Final deve ser **coluna+1**, pois a √∫ltima coluna informada ser√° desconsiderada;
    - Ser√° informada at√© a coluna anterior da coluna informada.

```python
# loc[Linhas, faixa √≠ndice colunas]
dados.iloc[10:20, 1:6]
```

```python
# Pegando informa√ß√µes da coluna informada at√© a √∫ltima
dados.iloc[10:20, 10: ]
```

```python
# Agora pegando dados da primeira coluna at√© a coluna informada
dados.iloc[10:20, :6]
```
