```python
import pandas as pd
```

```python
dados = pd.read_csv('train.csv')
```

```python
dados.head()
```

```python
# Exemplo: dados dos passageiros sobreviventes
dados.query('Survived == 1')
```

```python
# Exemplo: nomes dos sobreviventes
dados.query('Survived == 1')['Name']
```

```python
# Exemplo: nomes e idades dos sobreviventes
dados.query('Survived == 1')[['Name', 'Age']]
```

- TambÃ©m podemos fazer buscas por `strings` usando query.
- O termo da nossa busca deve ser buscado por aspas duplas (``""``).

```python
# Exemplo: dados de todas passageiras sobreviventes
dados.query('Sex == "female"')
```

```python
# Exemplo: todos os passageiros com idades a partir de 40 anos
dados.query('Age >= 40')
```

- Podemos realizar buscas com condiÃ§Ãµes mÃºltiplas em uma query.

```python
# Exemplo: Passageiras a partir de 40 anos
dados.query('Age >= 40 & Sex == "female"')
```

- Podemos criar variÃ¡veis para colocÃ¡-las em nossa query.
- Na query devemos informar o nome da variÃ¡vel inicializando com carectere "`@`".

```python
# Exemplo: Todos passageiros a partir de 30 anos
idade = 30
dados.query('Age >= @idade')
```
