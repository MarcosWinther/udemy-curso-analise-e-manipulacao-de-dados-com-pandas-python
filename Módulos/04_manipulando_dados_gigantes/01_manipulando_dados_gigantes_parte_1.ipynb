## M√©todo Chunksize
- No Pandas, o par√¢metro chunksize √© utilizado ao ler arquivos grandes, permitindo process√°-los em partes menores em vez de carreg√°-los completamente na mem√≥ria.
- Isso √© √∫til para lidar com grandes volumes de dados sem consumir muitos recursos do sistema.

### Como funciona?
- Quando voc√™ usa chunksize, o Pandas l√™ o arquivo em lotes (chunks), retornando um iterador que pode ser processado iterativamente.

### Vantagens do chunksize:
‚úî Permite processar arquivos grandes sem estourar a mem√≥ria RAM. <br>
‚úî √â √∫til para an√°lise de dados em lotes. <br>
‚úî Pode ser combinado com opera√ß√µes como concat() para montar o DataFrame completo, se necess√°rio.

```python
dfc = pd.read_csv('dataset_ENEM_2023/DADOS/MICRODADOS_ENEM_2023.csv', encoding='ISO-8859-1', sep=';', chunksize=1000000)
```
