## Manipulando Colunas

```python
# Informa todas as colunas
dados.columns
```

```python
type(dados['Survived'])
```

```python
# Criando uma cópia do meu DataFrame
dados_copia = dados.copy()

dados_copia
```

```python
dados_copia.rename(columns={'PassengerId': 'Id Passageiro', 'Survived': 'Sobreviveu', 'Pclass': 'Classe', 'Name': 'Nome', 'Sex': 'Sexo', 'Age': 'Idade',
                           'SibSp': 'Membros Família', 'Parch': 'Pais e Filhos a Bordo', 'Ticket': 'Bilhete', 'Fare': 'Valor Passage,', 
                            'Cabin': 'Cabine', 'Embarked': 'Portão de Embarque'}, inplace=True)
```

```python
dados_copia
```

```python
# Como é uma lista, podemos percorrer essa lista
for i in dados_copia.columns:
    print(i)
```

```python
# Criando outra cópia dos nossos dados para aprender excluir colunas
dados_copia2 = dados.copy()
```

```python
# Excluindo coluna utilizando o comando del
del dados_copia['Id Passageiro']
```

```python
dados_copia
```

```python
# Utilizando o método drop para excluir colunas
# dado.drop(['colunas'], axis= '0 para excluir linhas e 1 para excluir coluna inteira', inplace se for True excluirá de forma definitiva)
dados_copia2.drop(['PassengerId', 'Survived'], axis= 1, inplace= True)
```
